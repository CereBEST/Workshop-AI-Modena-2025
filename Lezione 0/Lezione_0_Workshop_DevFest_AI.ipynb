1. Alzatevi tutti in piedi e pensate al numero 1
2. Scegliete un solo compagno o una sola compagna vicino a voi (davanti, a sinistra, destra o dietro), sommate i numeri che avete e ricordatevi entrambi la somma

4. La persona rimasta in piedi torna allo step 2 poi esegue lo step 3 e poi via così fino a che non rimarrà in tutta la stanza una sola persona in piedi

# Machine Learning

Una macchina che viene *allenata* e che impara automaticamente le regole per risolvere un problema. Per farlo necessita di:
* Dati di input
* Esempi dell'output che ci aspettiamo
* Un modo per misurare se l'algoritmo sta facendo bene

Un modello di machine learning cerca quindi automaticamente regole e rappresentazioni utili dei dati di input, all'interno di uno spazio di possibilità predefinito, guidata da un feedback.

## Differenti tipi di Machine Learning

Gli algoritmi di Machine Learning possono essere suddivisi a seconda del loro stile di apprendimento in:
* Supervisionati (Supervised Learning)
* Non Supervisionati (Unsupervised Learning)
* Apprendimento per Rinforzo (Reinforcement Learning)

## Supervised Learning

* I dati usati per allenare il modello (training data) sono annotati con le etichette corrispondenti all'output corrispondente
* Le variabili di input sono chiamate in gergo *features*, mentre la variabile di output è chiamata *target*

## Unsupervised Learning

* Non vengono usati dati etichettati per l'addestramento dei modelli
* Il modello trova automaticamente degli schemi (pattern) o dei gruppi (clusters) all'interno del dataset

## Reinforcement Learning

* Una via di mezzo tra apprendimento supervisionato e non supervisionato
* Non utilizza etichette per allenare il modello. Tuttavia usa ricompense e punizioni per far si che il modello apprenda delle regole (policy) utili per massimizzare le ricompense

## Classificazione dei modelli in base al problema da risolvere

I modelli di Machine Learning possono diffrenziarsi anche per la tipologia di problema che sono chiamati a risolvere:
* Regressione: predizione di un valore numerico
* Classificazione: predizione di una classe o categoria

# Deep Learning

* Modelli (reti neurali) in cui le capacità di apprendimento sono ampie per la presenza di numerosi strati nella loro architettura che danno la possibilità di imparare molte più regole e rappresentazioni dei dati di input. "Quanti strati" rappresenta proprio la profontità (depth) del modello (deep o shallow).

* Le reti neurali possono al limite ispirarsi in alcuni aspetti al nostro cervello, ma non sono affatto una sua riproduzione in scala e non seguono le sue regole biologiche.

* Il Deep Learning è quindi una via multi-strato utile (multi-layer) per imparare una rappresentazione dei dati
* I dati passano attraverso dei "filtri" per poi uscirne "purificati"
* L'apprendimento di ogni modello di Deep Learning sta nei suoi pesi (weights)
* In altre parole la trasformazione dei dati di input in output è parametrizzata dai pesi di ogni strato

* La funzione di *loss* è adibita a misurare, durante l'apprendimento, quanto distante è l'output generato dal modello con quello reale (ground truth)
* Il valore di loss viene usato dal modello come feedback per aggiustare i pesi di poco per andare nella direzione che porta il modello a minimizzare il valore della loss. Questo processo di *ottimizzazione* numerica è regolato da un'algoritmo di *backpropagation*
* All'inizio dell'apprendimento il valore dei pesi viene assegnato randomicamente, dopodiché tramite questo meccanismo di apprendimento ripetuto x volte (epoche) i pesi assumono valori tali da minimizzare la funzione di loss

## Natural Language Processing (NLP)

* Area del Deep Learning che ha lo scopo di programmare macchine che sappiano comprendere e produrre linguaggio
* Tipici compiti del Natural Language Processing sono la traduzione di un testo tra lingue diverse, la comprensione del linguaggio naturale e la generazione di testo date alcune istruzioni

## Computer Vision (CV)

* Area del Deep Learning deputata alla programmazione di macchine che possono comprendere stimoli visivi e riprodurli
* Tipici compiti di Computer Vision sono la classificazione delle immagini, il riconoscimento degli oggetti e la loro localizzazione, la segmentazione degli oggetti in un'immagine, la stima della posa di persone o animali e la generazione di immagini o video

## You Only Look Once (YOLO)

* Una tipologia di modello di Computer Vision molto veloce utile per portare avanti tutti i task descritti in tempo reale
* Possiamo adattare e raffinare l'apprendimento di questa rete (fine tuning) attraverso un nostro dataset, ad esempio per riconoscere oggetti particolari o specifici
* Per riaddestrarlo non abbiamo bisogno di molti dati perché possiamo sfruttare le "conoscenze" generali che il modello ha su oggetti generici della nostra vita quotidiana su cui è stato già addestrato (transfer learning)

## Importo le librerie Python che mi servono

```python
from ultralytics import YOLO  # funzione utile per caricare modello di CV
import cv2  # OpenCV liberia utile per manipolazione delle immagini
from google.colab import files  # funzione per manipolare i files
from google.colab.patches import cv2_imshow  # funzione utile per viz img
```

## Carico le immagini che voglio analizzare, le rimpicciolisco del 50% e le visualizzo

```python
upload = files.upload()

LIST_IMG = [i for i in upload]

LIST_IMG_RES = []

for i in LIST_IMG:
  img = cv2.imread(i)
  img_res = cv2.resize(img, (0, 0), fx=0.5, fy=0.5)
  print(i)
  cv2_imshow(img_res)
  new_name = "res_" + i
  LIST_IMG_RES.append(new_name)
  cv2.imwrite(new_name, img_res)
```

## Classificazione delle immagini

```python
model_cls = YOLO("yolo11n-cls.pt")

predictions_cls = model_cls.predict(LIST_IMG_RES)

for p in predictions_cls:
  p.plot(), p.show()
```

## Object detection

```python
model_obj_det = YOLO("yolo11n.pt")

predictions_obj_det = model_obj_det.predict(LIST_IMG_RES)

for p in predictions_obj_det:
  p.plot(), p.show()
```

## Object segmentation

```python
model_seg = YOLO("yolo11n-seg.pt")

predictions = model_seg.predict(LIST_IMG_RES)

for p in predictions:
  p.plot(), p.show()
```

## Pose estimation

```python
model_pose = YOLO("yolo11x-pose.pt")

predictions_pose = model_pose.predict(LIST_IMG_RES)

for p in predictions_pose:
  p.plot(), p.show()
```

# YOLO tramite webcam live

```python
from IPython.display import display, Javascript
from google.colab.output import eval_js
from base64 import b64decode

def take_photo(filename='photo.jpg', quality=0.8):
  js = Javascript('''
    async function takePhoto(quality) {
      const div = document.createElement('div');
      const capture = document.createElement('button');
      capture.textContent = 'Capture';
      div.appendChild(capture);

      const video = document.createElement('video');
      video.style.display = 'block';
      const stream = await navigator.mediaDevices.getUserMedia({video: true});

      document.body.appendChild(div);
      div.appendChild(video);
      video.srcObject = stream;
      await video.play();

      // Resize the output to fit the video element.
      google.colab.output.setIframeHeight(document.documentElement.scrollHeight, true);

      // Wait for Capture to be clicked.
      await new Promise((resolve) => capture.onclick = resolve);

      const canvas = document.createElement('canvas');
      canvas.width = video.videoWidth;
      canvas.height = video.videoHeight;
      canvas.getContext('2d').drawImage(video, 0, 0);
      stream.getVideoTracks()[0].stop();
      div.remove();
      return canvas.toDataURL('image/jpeg', quality);
    }
    ''')
  display(js)
  data = eval_js('takePhoto({})'.format(quality))
  binary = b64decode(data.split(',')[1])
  with open(filename, 'wb') as f:
    f.write(binary)
  return filename
```

```python
from IPython.display import Image
try:
  filename = take_photo()
  print('Saved to {}'.format(filename))

  # Show the image which was just taken.
  display(Image(filename))
  print(filename)
  my_img = model_seg.predict(filename)
  for p in my_img:
    p.plot(), p.show()
except Exception as err:
  # Errors will be thrown if the user does not have a webcam or if they do not
  # grant the page permission to access it.
  print(str(err))
```
